```python
import requests
import bs4

def grabing_authors():
    list_of_autors = set()
    f = 1

    res = requests.get(f"https://quotes.toscrape.com/page/{f}/")
    soup = bs4.BeautifulSoup(res.text, "lxml")
    all_quotes = soup.select('.quote')

    while len(all_quotes) != 0:

        for i in range(len(all_quotes)):

            quote = all_quotes[i]
            autor = quote.select('.author')[0].text
            list_of_autors.add(autor)
        f += 1

        res = requests.get(f"https://quotes.toscrape.com/page/{f}/")
        soup = bs4.BeautifulSoup(res.text, "lxml")
        all_quotes = soup.select('.quote')

    return list_of_autors
```

```python
grabing_authors()
```
